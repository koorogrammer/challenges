# baseline
- 일단 연속이 아닌데 unique가 너무 많을 것 같은 Name, Ticket을 빼고 해봄
- automl을 이용해 base score와 잘 나오는 모델을 파악함

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']
label = 'Survived'
```

```python
from pycaret.classification import *
```

```python
s = setup(df[features + [label]], target=label)
```

```python
pd.set_option('display.max_columns', 200)
get_config('X_train').head(5)
# age랑 Fare 빼고는 다 one-hot 인코딩 되었음
```

```python
best = compare_models(sort='Accuracy')
```

```python
best
# best score 0.8154
```

```python
predictions = predict_model(best, data=test_df[features])
```

```python
submissions = pd.DataFrame({'PassengerId': test_df['PassengerId'], 'Survived': predictions['Label']})
submissions.to_csv('basline.csv', index=False)
```
